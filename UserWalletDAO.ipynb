```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class UserWalletBean:
    userWalletId = ""
    userId = ""
    walletBalance = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class UserWalletDAO:
    def addUserWallet(self, u=UserWalletBean()):
        try:
     
            sql = "INSERT INTO bp_user_wallets(user_wallet_id, user_id, wallet_balance, entry_date) VALUES(%s, %s, %s, %s)"
            data = (u.userWalletId, u.userId, u.walletBalance, u.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateUserWallet(self, u=UserWalletBean()):
        try:
            
            sql = "UPDATE bp_user_wallets SET user_id=%s, wallet_balance=%s, entry_date=%s WHERE user_wallet_id=%s"
            data = (u.userId, u.walletBalance, u.entryDate, u.userWalletId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
                      
    def deleteUserWallet(self, userWalletId):
        try:
            
            sql = "DELETE FROM bp_user_wallets WHERE user_wallet_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (userWalletId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getUserWallet(self, userWalletId):
        try:
            
            sql = "SELECT user_wallet_id Id, user_id user, wallet_balance balance, entry_date date FROM bp_user_wallets WHERE user_wallet_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (userWalletId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getUserWalletList(self):
        try:
            
            sql = "SELECT user_wallet_id Id, user_id user, wallet_balance balance, entry_date date FROM bp_user_wallets"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```

```python
#Testing DAO
udao=UserWalletDAO()


def testAddMethod():
    
    u = UserWalletBean()
    
    u.userWalletId  = "UW03"
    u.userId = "U001"
    u.walletBalance = 50.0
    u.entryDate = "2020-05-09"
    
    r = udao.addUserWallet(u)
    print("Response: ",r)

def testUpdateMethod():
    
    u = UserWalletBean()
    
    u.userId = "U002"
    u.walletBalance = 1000.0
    u.entryDate = "2020-05-09"
    u.userWalletId  = "UW02"
    
    r = udao.updateUserWallet(u)
    print("Response: ",r)

def testDeleteMethod():
    
    testId = "UW03"

    r = udao.deleteUserWallet(testId)
    print("Response: ",r)

def testGetRecordMethod():
    
    testId = "UW01"
    record = udao.getUserWallet(testId)
    print(record)
    
def testGetRecordListMethod():
    
    records = udao.getUserWalletList()
    print(records)

# testAddMethod()
# testUpdateMethod()
# testDeleteMethod()
testGetRecordMethod()
testGetRecordListMethod()
```
