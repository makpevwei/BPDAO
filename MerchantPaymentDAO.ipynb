```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class MerchantPaymentBean:
    paymentId = ""
    userId = ""
    merchantCode = ""
    serviceDescription = ""
    paymentDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class MerchantPaymentDAO:
    def addMerchantPayment(self, m=MerchantPaymentBean()):
        try:
     
            sql = "INSERT INTO bp_merchant_payments(payment_id, user_id, merchant_code, description_of_service, payment_date) VALUES(%s, %s, %s, %s, %s)"
            data = (m.paymentId, m.userId, m.merchantCode, m.serviceDescription, m.paymentDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateMerchantPayment(self, m=MerchantPaymentBean()):
        try:
            
            sql = "UPDATE bp_merchant_payments SET user_id=%s, merchant_code=%s, description_of_service=%s, payment_date=%s WHERE payment_id=%s"
            data = (m.userId, m.merchantCode, m.serviceDescription, m.paymentDate, m.paymentId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
                      
    def deleteMerchantPayment(self, paymentId):
        try:
            
            sql = "DELETE FROM bp_merchant_payments WHERE payment_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (paymentId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getMerchantPayment(self, paymentId):
        try:
            
            sql = "SELECT payment_id Id, user_id user, merchant_code merchant, description_of_service description, payment_date date FROM bp_merchant_payments WHERE payment_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (paymentId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getMerchantPaymentList(self):
        try:
            
            sql = "SELECT payment_id Id, user_id user, merchant_code merchant, description_of_service description, payment_date date FROM bp_merchant_payments"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
