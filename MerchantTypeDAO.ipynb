```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class MerchantTypeBean:
    merchantTypeId = ""
    merchantType = ""
    description = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class MerchantTypeDAO:
    def addMerchantType(self, m=MerchantTypeBean()):
        try:
     
            sql = "INSERT INTO bp_merchant_types(merchant_type_id, merchant_type, description,entry_date) VALUES(%s, %s, %s, %s)"
            data = (m.merchantTypeId, m.merchantType, m.description, m.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateMerchantType(self, m=MerchantTypeBean()):
        try:
            
            sql = "UPDATE bp_merchant_types SET merchant_type=%s, description=%s, entry_date=%s WHERE merchant_type_id=%s"
            data =  (m.merchantType, m.description, m.entryDate, m.merchantTypeId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            

            
    def deleteMerchantType(self, merchantTypeId):
        try:
            
            sql = "DELETE FROM bp_merchant_types WHERE merchant_type_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (merchantTypeId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getMerchantType(self, merchantTypeId):
        try:
            
            sql = "SELECT merchant_type_id id, merchant_type type, description description,  entry_date date FROM bp_merchant_types WHERE merchant_type_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (merchantTypeId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getMerchantTypeList(self):
        try:
            
            sql = "SELECT merchant_type_id id, merchant_type type, description description,  entry_date date FROM bp_merchant_types"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```

```python
#Testing DAO
mtdao=MerchantTypeDAO()


def testAddMethod():
    
    m=MerchantTypeBean
    m.merchantTypeId = "MT03"
    m.merchantType = "Bank"
    m.description = "Bank"
    m.entryDate = "2020-05-06"

    r = mtdao.addMerchantType(m)
    print("Response: ",r)

def testUpdateMethod():
    
    m=MerchantTypeBean
    m.merchantType = "Stock Exchange"
    m.description = "Stock Exchange"
    m.entryDate = "2020-05-06"
    m.merchantTypeId = "MT02"

    r = mtdao.updateMerchantType(m)
    print("Response: ",r)

def testDeleteMethod():
    
    testId = "MT03"

    r = mtdao.deleteMerchantType(testId)
    print("Response: ",r)

def testGetRecordMethod():
    
    testId = "MT01"
    record = mtdao.getMerchantType(testId)
    print(record)
    
def testGetRecordListMethod():
    
    records = mtdao.getMerchantTypeList()
    print(records)

# testAddMethod()
# testUpdateMethod()
# testDeleteMethod()
testGetRecordMethod()
testGetRecordListMethod()
```
