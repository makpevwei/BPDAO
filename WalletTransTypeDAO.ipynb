```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class WalletTransTypeBean:
    walletTransTypeid = ""
    walletTranstype = ""
    description = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class WalletTransTypeDAO:
    def addWalletTransType(self, w=WalletTransTypeBean()):
        try:
     
            sql = "INSERT INTO bp_wallet_trans_types(wallet_trans_type_id, wallet_trans_type, description, entry_date) VALUES(%s, %s, %s, %s)"
            data = (w.walletTransTypeid, w.walletTranstype, w.description, w.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateWalletTransType(self, w=WalletTransTypeBean()):
        try:
            
            sql = "UPDATE bp_wallet_trans_types SET wallet_trans_type=%s, description=%s, entry_date=%s WHERE wallet_trans_type_id=%s"
            data = (w.walletTranstype, w.description, w.entryDate, w.walletTransTypeid,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
                      
    def deleteWalletTransType(self, walletTransTypeid):
        try:
            
            sql = "DELETE FROM bp_wallet_trans_types WHERE wallet_trans_type_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (walletTransTypeid,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getWalletTransType(self, walletTransTypeid):
        try:
            
            sql = "SELECT wallet_trans_type_id Id, wallet_trans_type type, description description, entry_date date FROM bp_wallet_trans_types WHERE wallet_trans_type_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (walletTransTypeid,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getWalletTransTypeList(self):
        try:
            
            sql = "SELECT wallet_trans_type_id Id, wallet_trans_type type, description description, entry_date date FROM bp_wallet_trans_types"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```

```python
#Testing DAO
wdao=WalletTransTypeDAO()


def testAddMethod():
    
    w = WalletTransTypeBean()
    
    w.walletTransTypeid = "WT03"
    w.walletTranstype = "Banking"
    w.description = "Banking"
    w.entryDate = "2020-05-09"
 
    r = wdao.addWalletTransType(w)
    print("Response: ",r)

def testUpdateMethod():
    
    w = WalletTransTypeBean()
    
    w.walletTranstype = "Stock"
    w.description = "Stock"
    w.entryDate = "2020-05-09"
    w.walletTransTypeid = "WT02"
    
    r = wdao.updateWalletTransType(w)
    print("Response: ",r)

def testDeleteMethod():
    
    testId = "WT03"

    r = wdao.deleteWalletTransType(testId)
    print("Response: ",r)

def testGetRecordMethod():
    
    testId = "WT01"
    record = wdao.getWalletTransType(testId)
    print(record)
    
def testGetRecordListMethod():
    
    records = wdao.getWalletTransTypeList()
    print(records)

# testAddMethod()
# testUpdateMethod()
# testDeleteMethod()
testGetRecordMethod()
testGetRecordListMethod()
```
