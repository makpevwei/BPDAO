```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class MerchantBean:
    merchantCode = ""
    merchantTypeId = ""
    branchCode = ""
    merchantName = ""
    rc = ""
    url = ""
    location = ""
    longitude = ""
    latitude = ""
    address = ""
    mobile = ""
    altMobile = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class MerchantDAO:
    def addMerchant(self, m=MerchantBean()):
        try:
     
            sql = "INSERT INTO bp_merchants(merchant_code, merchant_type_id, branch_code, merchant_name, rc_rn_no, merchant_url, merchant_location, longitude, latitude, merchant_address, mobile_no, alt_mobile_no, entry_date) VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
            data = (m.merchantCode, m.merchantTypeId, m.branchCode, m.merchantName, m.rc, m.url, m.location, m.longitude, m.latitude, m.address, m.mobile, m.altMobile, m.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateMerchant(self, m=MerchantBean()):
        try:
            
            sql = "UPDATE bp_merchants SET merchant_type_id=%s, branch_code=%s, merchant_name=%s, rc_rn_no=%s, merchant_url=%s, merchant_location=%s, longitude=%s, latitude=%s, merchant_address=%s, mobile_no=%s, alt_mobile_no=%s, entry_date=%s WHERE merchant_code=%s"
            data = (m.merchantTypeId, m.branchCode, m.merchantName, m.rc, m.url, m.location, m.longitude, m.latitude, m.address, m.mobile, m.altMobile, m.entryDate, m.merchantCode,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
                      
    def deleteMerchant(self, merchantCode):
        try:
            
            sql = "DELETE FROM bp_merchants WHERE merchant_code=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (merchantCode,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getMerchant(self, merchantCode):
        try:
            
            sql = "SELECT merchant_code code, merchant_type_id type, branch_code branch, merchant_name name, rc_rn_no rc, merchant_url url, merchant_location location, longitude longitude, latitude latitude, merchant_address address, mobile_no mobile, alt_mobile_no alt, entry_date date FROM bp_merchants WHERE merchant_code=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (merchantCode,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getMerchantList(self):
        try:
            
            sql = "SELECT merchant_code code, merchant_type_id type, branch_code branch, merchant_name name, rc_rn_no rc, merchant_url url, merchant_location location, longitude longitude, latitude latitude, merchant_address address, mobile_no mobile, alt_mobile_no alt, entry_date date FROM bp_merchants"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
