```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class UserBean:
    userId = ""
    userRoleId = ""
    uPassword = ""
    firstName = ""
    lastName = ""
    gender = ""
    mobile = ""
    userStatus = ""
    lastLogin = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class UserDAO:
    def addUser(self, u=UserBean()):
        try:
     
            sql = "INSERT INTO bp_users(user_id, user_role_id, u_password, first_name, last_name, gender, mobile_no, user_status, last_login, entry_date) VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
            data = (u.userId, u.userRoleId, u.uPassword, u.firstName, u.lastName, u.gender, u.mobile, u.userStatus, u.lastLogin, u.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateUser(self, u=UserBean()):
        try:
            
            sql = "UPDATE bp_users SET user_role_id=%s, u_password=%s, first_name=%s, last_name=%s, gender=%s, mobile_no=%s, user_status=%s, last_login=%s, entry_date=%s WHERE user_id=%s"
            data = (u.userRoleId, u.uPassword, u.firstName, u.lastName, u.gender, u.mobile, u.userStatus, u.lastLogin, u.entryDate, u.userId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
                      
    def deleteUser(self, userId):
        try:
            
            sql = "DELETE FROM bp_users WHERE user_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (userId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getUser(self, userId):
        try:
            
            sql = "SELECT user_id Id, user_role_id role, u_password password, first_name first, last_name last, gender gender, mobile_no mobile, user_status status, last_login login, entry_date date FROM bp_users WHERE user_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (userId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getUserList(self):
        try:
            
            sql = "SELECT user_id Id, user_role_id role, u_password password, first_name first, last_name last, gender gender, mobile_no mobile, user_status status, last_login login, entry_date date FROM bp_users"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
