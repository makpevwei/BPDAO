```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class UserBeneficiaryBean:
    userMerchantId = ""
    userId = ""
    merchantCode = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class UserBeneficiaryDAO:
    def addUserBeneficiary(self, u=UserBeneficiaryBean()):
        try:
     
            sql = "INSERT INTO bp_user_beneficiaries(user_merchant_id, user_id, merchant_code, entry_date) VALUES(%s, %s, %s, %s)"
            data = (u.userMerchantId, u.userId, u.merchantCode, u.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateUserBeneficiary(self, u=UserBeneficiaryBean()):
        try:
            
            sql = "UPDATE bp_user_beneficiaries SET user_id=%s, merchant_code=%s, entry_date=%s WHERE user_merchant_id=%s"
            data = (u.userId, u.merchantCode, u.entryDate, u.userMerchantId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
                      
    def deleteUserBeneficiary(self, userMerchantId):
        try:
            
            sql = "DELETE FROM bp_user_beneficiaries WHERE user_merchant_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (userMerchantId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getUserBeneficiary(self, userMerchantId):
        try:
            
            sql = "SELECT user_merchant_id Id, user_id user, merchant_code merchant, entry_date date FROM bp_user_beneficiaries WHERE user_merchant_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (userMerchantId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getUserBeneficiaryList(self):
        try:
            
            sql = "SELECT user_merchant_id Id, user_id user, merchant_code merchant, entry_date date FROM bp_user_beneficiaries"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
