```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class UserRoleBean:
    userRoleId = ""
    userRole = ""
    description = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class UserRoleDAO:
    def addUserRole(self, u=UserRoleBean()):
        try:
     
            sql = "INSERT INTO bp_user_roles(user_role_id, user_role, user_role_desc, entry_date) VALUES(%s, %s, %s, %s)"
            data = (u.userRoleId, u.userRole, u.description, u.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            
    def updateUserRole(self, u=UserRoleBean()):
        try:
            
            sql = "UPDATE bp_user_roles SET user_role=%s, user_role_desc=%s, entry_date=%s WHERE user_role_id=%s"
            data = (u.userRole, u.description, u.entryDate, u.userRoleId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

            
    def deleteUserRole(self, userRoleId):
        try:
            
            sql = "DELETE FROM bp_user_roles WHERE user_role_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (userRoleId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getUserRole(self, userRoleId):
        try:
            
            sql = "SELECT user_role_id Id, user_role role, user_role_desc description, entry_date date FROM bp_user_roles WHERE user_role_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (userRoleId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getUserRoleList(self):
        try:
            
            sql = "SELECT user_role_id Id,user_role role, user_role_desc description, entry_date date FROM bp_user_roles"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
