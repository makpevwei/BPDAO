```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class BranchBean:
    branchCode = ""
    bankCode = ""
    branch = ""
    location = ""
    contactPerson = ""
    contactMobile = ""
    contactEmail = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class BranchDAO:
    def addBranch(self, b=BranchBean()):
        try:
     
            sql = "INSERT INTO bp_bank_branches(branch_code, bank_code, branch, location, contact_person, contact_mobile, contact_email, entry_date) VALUES(%s, %s, %s, %s, %s, %s, %s, %s)"
            data = (b.branchCode, b.bankCode, b.branch, b.location, b.contactPerson, b.contactMobile, b.contactEmail, b.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateBranch(self, b=BranchBean()):
        try:
            
            sql = "UPDATE bp_bank_branches SET bank_code=%s, branch=%s, location=%s, contact_person=%s, contact_mobile=%s, contact_email=%s, entry_date=%s WHERE branch_code=%s"
            data = (b.bankCode, b.branch, b.location, b.contactPerson, b.contactMobile, b.contactEmail, b.entryDate, b.branchCode,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
                      
    def deleteBranch(self, branchCode):
        try:
            
            sql = "DELETE FROM bp_bank_branches WHERE branch_code=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (branchCode,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getBranch(self, branchCode):
        try:
            
            sql = "SELECT branch_code code, bank_code bank, branch branch, location location, contact_person person, contact_mobile mobile, contact_email email, entry_date date FROM bp_bank_branches WHERE branch_code=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (branchCode,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getBranchList(self):
        try:
            
            sql = "SELECT branch_code code, bank_code bank, branch branch, location location, contact_person person, contact_mobile mobile, contact_email email, entry_date date FROM bp_bank_branches"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
