```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class BankBean:
    bankCode = ""
    bank = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class BankDAO:
    def addBank(self, b=BankBean()):
        try:
     
            sql = "INSERT INTO bp_banks(bank_code, bank,entry_date) VALUES(%s, %s, %s)"
            data = (b.bankCode, b.bank, b.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateBank(self, b=BankBean()):
        try:
            
            sql = "UPDATE bp_banks SET bank=%s, entry_date=%s WHERE bank_code=%s"
            data =  (b.bank, b.entryDate, b.bankCode,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            

            
    def deleteBank(self, bankCode):
        try:
            
            sql = "DELETE FROM bp_banks WHERE bank_code=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (bankCode,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getBank(self, bankCode):
        try:
            
            sql = "SELECT bank_code code, bank bank, entry_date date FROM bp_banks WHERE bank_code=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (bankCode,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getBankList(self):
        try:
            
            sql = "SELECT bank_code code, bank bank, entry_date date FROM bp_banks"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```

```python
#Testing DAO
bdao=BankDAO()


def testAddMethod():
    
    b=BankBean
    b.bankCode = "B03"
    b.bank = "GTB"
    b.entryDate = "2020-05-06"

    r = bdao.addBank(b)
    print("Response: ",r)

def testUpdateMethod():
    
    b=BankBean
    b.bank = "FBN"
    b.entryDate = "2020-05-06"
    b.bankCode = "B02"

    r = bdao.updateBank(b)
    print("Response: ",r)

def testDeleteMethod():
    
    testId = "B03"

    r = bdao.deleteBank(testId)
    print("Response: ",r)

def testGetRecordMethod():
    
    testId = "B01"
    record = bdao.getBank(testId)
    print(record)
    
def testGetRecordListMethod():
    
    records = bdao.getBankList()
    print(records)

# testAddMethod()
# testUpdateMethod()
# testDeleteMethod()
testGetRecordMethod()
testGetRecordListMethod()
```
