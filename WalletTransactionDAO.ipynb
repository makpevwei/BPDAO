```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class WalletTransactionBean:
    transactionId = ""
    userWalletId = ""
    walletTransType = ""
    transactionAmount = ""
    transactionDate =  datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class WalletTransactionDAO:
    def addWalletTransaction(self, w=WalletTransactionBean()):
        try:
     
            sql = "INSERT INTO bp_wallet_transactions(transaction_id, user_wallet_id, wallet_trans_type_id, transaction_amount, transaction_date, entry_date) VALUES(%s, %s, %s, %s, %s, %s)"
            data = (w.transactionId, w.userWalletId, w.walletTransType, w.transactionAmount, w.transactionDate, w.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateWalletTransaction(self, w=WalletTransactionBean()):
        try:
            
            sql = "UPDATE bp_wallet_transactions SET user_wallet_id=%s, wallet_trans_type_id=%s, transaction_amount=%s, transaction_date=%s, entry_date=%s WHERE transaction_id=%s"
            data = (w.userWalletId, w.walletTransType, w.transactionAmount, w.transactionDate, w.entryDate, w.transactionId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
                      
    def deleteWalletTransaction(self, transactionId):
        try:
            
            sql = "DELETE FROM bp_wallet_transactions WHERE transaction_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (transactionId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getWalletTransaction(self, transactionId):
        try:
            
            sql = "SELECT transaction_id Id, user_wallet_id walletId, wallet_trans_type_id transId, transaction_amount amount, transaction_date transDate, entry_date date FROM bp_wallet_transactions WHERE transaction_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (transactionId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getWalletTransactionList(self):
        try:
            
            sql = "SELECT transaction_id Id, user_wallet_id walletId, wallet_trans_type_id transId, transaction_amount amount, transaction_date transDate, entry_date date FROM bp_wallet_transactions"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
